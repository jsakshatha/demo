# Statistics Overview

Statistics is the branch of mathematics that deals with the collecting, organizing, presenting, analyzing, and interpreting of data. These steps work together to turn raw information into useful knowledge.

## Steps in Statistics

### Collecting Data
Gathering relevant information (e.g., asking students their test scores).

### Organizing Data
Arranging the data systematically (e.g., making tables or sorting scores).

### Presenting Data
Showing the data clearly using tables, charts, or graphs so others can understand it.

### Analyzing Data
Calculating summaries like averages and finding patterns.

### Interpreting Data
Explaining what the analysis means and drawing conclusions.

# Population and Sample in Statistics

## Population

- Population refers to the entire group or set of individuals, objects, or data points that you want to study or get information about.
- It includes every member of that group.
- The characteristics of a population (like the exact mean or standard deviation) are called **parameters**.
- **Examples:**  
  - All students in a school.  
  - Every book in a library.  
  - All voters in a city.  

## Sample

- A sample is a smaller subset taken from the population.  
- It is used when it's impractical or impossible to collect data from the entire population.  
- The characteristics of a sample (like the sample mean or sample standard deviation) are called **statistics**.  
- Samples are used to make estimates or inferences about the whole population.  
- **Examples:**  
  - 50 students selected randomly from a school.  
  - 100 books randomly checked in a library.  
  - 500 voters questioned in a city survey.  

---

### Key Differences

| Feature           | Population                              | Sample                                   |
|-------------------|---------------------------------------|-----------------------------------------|
| Definition        | Entire group of interest               | Subset of the population                 |
| Size              | Usually large or complete group        | Smaller part of the population            |
| Data Type         | Parameters (true population values)   | Statistics (estimates from sample)       |
| Data Collection   | Census (complete enumeration)          | Sampling (partial data collection)       |
| Use               | Gives exact information                | Used to make inferences about population |

---

### Why Use Samples?

- Populations can be too large or difficult to study completely.
- Sampling saves time and resources.
- Proper sampling methods allow us to make reliable conclusions about the population using sample data.

---

This explanation helps differentiate population and sample clearly for better understanding in statistics.

## üî∑ 1. Probability Sampling Methods


### üü¢ a) Simple Random Sampling
- **Explanation:** Every student has an equal chance of being selected.
- **Example:** From a list of 500 students, randomly pick 50 using a lottery method or random number generator.

---


### üü¢ b) Systematic Sampling
- **Explanation:** Select every *k-th* student from a list.
- **Example:** Arrange all 500 students alphabetically. Pick every 10th student (10th, 20th, 30th...).

---

### üü¢ c) Stratified Sampling
- **Explanation:** Divide the population into subgroups (strata), then randomly sample from each.
- **Example:** Divide students into 3 courses ‚Äî Science, Commerce, and Arts. Randomly pick 20 from each course.

---

### üü¢ d) Cluster Sampling
- **Explanation:** Divide population into clusters, then randomly choose entire clusters.
- **Example:** Out of 10 colleges in your city, randomly select 3 colleges. Survey **all students** in those colleges.

# üìä Measures of Central Tendency

---

## üß† 1. Definition

**Measures of Central Tendency** are statistical values that describe the **center or typical value** of a dataset.  
They indicate where most of the data points lie ‚Äî representing the ‚Äúaverage‚Äù value around which data is distributed.

The **three main measures** are:
1. **Mean**
2. **Median**
3. **Mode**

---

## üîπ 2. Mean (Arithmetic Average)

### üìò Definition:
The **mean** is the sum of all data values divided by the number of observations.

$$
\bar{x} = \frac{\sum_{i=1}^{n} x_i}{n}
$$

### üìä Example:
Dataset: [10, 20, 30, 40, 50]

$$
\bar{x} = \frac{10 + 20 + 30 + 40 + 50}{5} = 30
$$

---

## üîπ 3. Median (Middle Value)

### üìò Definition:
The **median** is the middle value when data is arranged in ascending or descending order.

- If **n** is odd ‚Üí middle value  
- If **n** is even ‚Üí average of the two middle values

### üìä Example:
Dataset: [5, 10, 15, 20, 25, 30]

Since 6 (even number of values):
$$
\text{Median} = \frac{15 + 20}{2} = 17.5
$$

---

## üîπ 4. Mode (Most Frequent Value)

### üìò Definition:
The **mode** is the data value that occurs **most frequently**.

### üìä Example:
Dataset: [2, 3, 3, 5, 7, 7, 7, 8, 10]

$$
\text{Mode} = 7
$$

#  Measures of Variability (Dispersion)

---

##  1. Definition

**Measures of Variability** (also called **Measures of Dispersion**) describe **how spread out or scattered** the data values are around the center (mean or median).

> In short, variability tells us **how consistent or inconsistent** the data is.

When variability is:
- **Low:** data points are close together (more consistent)
- **High:** data points are spread far apart (less consistent)

---

## üìö 2. Common Measures of Variability

| Measure | Description | Formula (Population) | Formula (Sample) |
|----------|--------------|----------------------|------------------|
| **Range** | Difference between the highest and lowest value | $$ R = X_{max} - X_{min} $$ | $$ R = X_{max} - X_{min} $$ |
| **Variance** | Average of squared deviations from the mean | $$ \sigma^2 = \frac{\sum (x_i - \mu)^2}{N} $$ | $$ s^2 = \frac{\sum (x_i - \bar{x})^2}{n - 1} $$ |
| **Standard Deviation (SD)** | Square root of variance | $$ \sigma = \sqrt{\frac{\sum (x_i - \mu)^2}{N}} $$ | $$ s = \sqrt{\frac{\sum (x_i - \bar{x})^2}{n - 1}} $$ |
| **Interquartile Range (IQR)** | Range between 75th and 25th percentile | $$ IQR = Q_3 - Q_1 $$ | $$ IQR = Q_3 - Q_1 $$ |

---


#  Range 

##  What is Range?
The **range** of a dataset is the **simplest measure of spread (variability)**.  

$$
\text{Range} = \text{Maximum Value} - \text{Minimum Value}
$$


---

##  Example with Dataset
Dataset:  
$
[12, 18, 25, 26, 30, 34, 40, 45, 50]
$

- **Minimum (min)** = 12  
- **Maximum (max)** = 50  

$$
\text{Range} = 50 - 12 = 38
$$



---

##  How to Interpret Range
`- A **small range** ‚Üí all values are close together.  
- A **large range** ‚Üí data is more spread out.  
- ‚ö†Ô∏è **Range depends only on 2 values (min & max)**, so it is very sensitive to **outliers**.  

**Example:**  
If one extra score **= 100** is added ‚Üí  
$
\text{New Range} = 100 - 12 = 88
$
This makes the spread look huge, even though most values are still between 12 and 50.

---

---

##  Variance Formula (Simplified)

$$
\text{Variance} = \frac{1}{N} \sum (x_i - \mu)^2
$$

- $ x_i $ : each data value  
- $ \mu $: mean  
-  N : number of data points

 It's the **average of the squared differences** from the mean.

---

##  What is Standard Deviation?

**Standard Deviation (SD)** is the **square root of variance**.

> It tells you **how much the data varies from the mean** in the **original units** (like marks, rupees, etc.).





$$
\text{Standard Deviation} = \sqrt{\text{Variance}}
$$

---

---
## Example: score={10,20,30,40,50}
## üîπ Step 1: Calculate the Mean (Œº)

$$
\text{Mean} = \frac{10 + 20 + 30 + 40 + 50}{5} = \frac{150}{5} = 30
$$

 So, the average score is **30**.

---

## üîπ Step 2: Find Deviations from the Mean

| Score (X) | Deviation (X ‚àí Œº) | Squared Deviation \((X - Œº)^2\) |
|-----------|--------------------|----------------------------------|
| 10        | 10 ‚àí 30 = -20       | (-20)¬≤ = 400                     |
| 20        | 20 ‚àí 30 = -10       | (-10)¬≤ = 100                     |

| 50        | 50 ‚àí 30 = 20        | 20¬≤ = 400                        |

---

## üîπ Step 3: Calculate Variance (œÉ¬≤)

$$
\text{Variance} = \frac{\sum (X - Œº)^2}{N} = \frac{400 + 100 + 0 + 100 + 400}{5} = \frac{1000}{5} = 200
$$

üìå Variance tells us how **spread out** the data is (in squared units).

---

## üîπ Step 4: Calculate Standard Deviation (œÉ)

$$
\text{Standard Deviation} = \sqrt{\text{Variance}} = \sqrt{200} ‚âà 14.14
$$

üìå This means, on average, each score is about **14.14 points away** from the mean.

---
This way you‚Äôd see that:

Values (10 and 50) are 20 away from the mean (biggest deviations).

Values (20 and 40) are 10 away from the mean.

 these deviations give you the variance = 200 and SD ‚âà 14.14.


- **Mean** is the average score: 30  
- **Variance** shows the average of the squared differences from the mean: 200  
- **Standard Deviation** is the square root of variance: ‚âà 14.14  


- A **low standard deviation** means scores are tightly clustered around the mean.

- It helps us understand where a particular data point stands in the dataset.

# üß† Machine Bolt Diameter Test

---

## üéØ Problem
A machine produces **bolts** with a **target diameter = 5 mm**.  
We want to check if the **average diameter has increased**.

---

# üîπ Step-by-Step Summary

| ü™ú **Step** | üßæ **Description** | üí¨ **Explanation / Example** |
|:------------|:------------------|:-----------------------------|
| **1Ô∏è‚É£ State the claim** | Machine makes bolts with mean diameter **5 mm** | Target (Œº‚ÇÄ) = **5 mm** |
| **2Ô∏è‚É£ Define hypotheses** | - **Null (H‚ÇÄ): Œº = 5 mm**  <br> - **Alternative (H‚ÇÅ): Œº > 5 mm** | We test if the mean **has increased** |
| **3Ô∏è‚É£ Collect sample data** | Measure some sample bolts (e.g., 10 pieces) | Example: 5.1, 5.0, 5.2, 5.3 ‚Ä¶ |
| **4Ô∏è‚É£ Calculate sample mean (xÃÑ)** | Find the average of sample diameters | Suppose xÃÑ = **5.15 mm** |
| **5Ô∏è‚É£ Perform test** | Use **t-test** or **z-test** | Formula:<br>  t = (xÃÑ ‚àí Œº‚ÇÄ) / (s / ‚àön) |
| **6Ô∏è‚É£ Set significance level (Œ±)** | Choose **Œ± = 0.05** | 5% chance of error |
| **7Ô∏è‚É£ Compare & conclude** | If **p-value < Œ±**, reject H‚ÇÄ | Example: p = 0.02 ‚Üí Mean has increased ‚úÖ |

---

# üß© Example Result

| **Step** | **Result** |
|:-----------|:-------------|
| **Null Hypothesis (H‚ÇÄ)** | Œº = 5 mm |
| **Alternative Hypothesis (H‚ÇÅ)** | Œº > 5 mm |

| **Conclusion** | ‚úÖ **Reject H‚ÇÄ ‚Üí Mean diameter has increased** |

---

# ‚ú® Summary
‚úÖ This is a **One-Tailed Test (Right Tail)**

#  1. **Independent t-test Example (Unpaired Groups)**

---

##  Scenario

We want to compare **marks of boys and girls** in a math test.

| **Group** | **Scores** |
|:-----------|:-------------|
| **Boys** | [55, 60, 65, 70, 75] |
| **Girls** | [65, 70, 72, 68, 74] |

---

##  Objective
Check if there is a **significant difference** between the **average marks** of boys and girls.

---

# üîπ Step-by-Step Summary

| ü™ú **Step** | üßæ **Description** | üí¨ **Explanation / Example** |
|:------------|:------------------|:-----------------------------|
| **1Ô∏è‚É£ State the claim** | Compare mean marks of **boys** and **girls** | Two independent groups |
| **2Ô∏è‚É£ Define hypotheses** | - **H‚ÇÄ:** Œº‚ÇÅ = Œº‚ÇÇ (means are equal) <br> - **H‚ÇÅ:** Œº‚ÇÅ ‚â† Œº‚ÇÇ (means are different) | We test if there is any difference |
| **3Ô∏è‚É£ Collect data** | Boys: [55, 60, 65, 70, 75] <br> Girls: [65, 70, 72, 68, 74] | Two separate samples |
| **4Ô∏è‚É£ Calculate sample means** | Boys avg = **65.0** <br> Girls avg = **69.8** | Girls have slightly higher marks |
| **5Ô∏è‚É£ Perform t-test** | Use **independent t-test formula** to compare means | t = -2.23 |
| **6Ô∏è‚É£ Set significance level (Œ±)** | Œ± = **0.05** | 5% chance of being wrong |
| **7Ô∏è‚É£ Compute p-value** | p = **0.056** | Compare this with Œ± |
| **8Ô∏è‚É£ Draw conclusion** | Since **p (0.056) > Œ± (0.05)** ‚Üí ‚ùå **Fail to Reject H‚ÇÄ** | There is **no significant difference** between boys‚Äô and girls‚Äô marks |

---

# ‚ú® Summary
‚úÖ Test Type: **Independent (Unpaired) t-test**  
‚úÖ Groups: **Boys vs Girls**  
‚úÖ p = 0.056 > 0.05 ‚Üí ‚ùå **Fail to Reject H‚ÇÄ**  
‚úÖ **Conclusion:** The difference in average marks is **not statistically significant**

## üéØ Situation

|            | **Science** | **Arts** | **Total** |

| **Girls**  | 10 | 30 | 40 |
| **Total**  | 40 | 40 | 80 |

## üéØ Objective

# üîπ Step-by-Step Summary

| ü™ú **Step** | üßæ **Description** | üí¨ **Explanation / Example** |
|:------------|:------------------|:-----------------------------|
| **1Ô∏è‚É£ State the claim** | We test if **gender** and **subject preference** are **independent** | Boys & girls choose subjects differently? |
| **2Ô∏è‚É£ Define hypotheses** | - **H‚ÇÄ:** Gender and subject preference are **independent** <br> - **H‚ÇÅ:** Gender and subject preference are **not independent** | Two-way relationship test |
| **3Ô∏è‚É£ Write observed frequencies (O)** | Use the given data table | e.g., O‚ÇÅ‚ÇÅ = 30, O‚ÇÅ‚ÇÇ = 10, etc. |
| **4Ô∏è‚É£ Compute expected frequencies (E)** | E = (Row Total √ó Column Total) / Grand Total | For Boys‚ÄìScience: (40√ó40)/80 = 20 |
| **5Ô∏è‚É£ Calculate all expected (E)** | Boys‚ÄìScience = 20 <br> Boys‚ÄìArts = 20 <br> Girls‚ÄìScience = 20 <br> Girls‚ÄìArts = 20 | Expected table: all = 20 |
| **6Ô∏è‚É£ Apply Chi-Square formula** | œá¬≤ = Œ£((O‚àíE)¬≤ / E) | = ((30‚àí20)¬≤/20) + ((10‚àí20)¬≤/20) + ((10‚àí20)¬≤/20) + ((30‚àí20)¬≤/20) |
| **7Ô∏è‚É£ Compute œá¬≤ value** | œá¬≤ = 10 + 5 + 5 + 10 = **30** | Calculated test statistic |
| **8Ô∏è‚É£ Find degrees of freedom (df)** | df = (r‚àí1)(c‚àí1) = (2‚àí1)(2‚àí1) = 1 | One degree of freedom |
| **9Ô∏è‚É£ Set significance level (Œ±)** | Œ± = **0.05** | 5% risk of error |
| **üîü Compare with critical value or p-value** | For df=1, œá¬≤‚ÇÄ.‚ÇÄ‚ÇÖ = 3.84 | œá¬≤ = 30 > 3.84 |
| **‚úÖ Draw conclusion** | Since œá¬≤ > 3.84 ‚Üí **Reject H‚ÇÄ** | ‚úÖ **Gender and subject preference are related** |
